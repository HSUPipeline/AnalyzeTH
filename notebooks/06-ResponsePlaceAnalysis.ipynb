```python
import numpy as np

from convnwb.io import load_nwbfile

from spiketools.plts.spatial import plot_positions
```

```python
import seaborn as sns
sns.set_context('talk')
```

## Settings

```python
# Define the base data folder
data_folder = '/Users/tom/Data/JacobsLab/WVTH/NWB'
#data_folder = '/scratch/tom.donoghue/TH/NWB'
```

```python
# Define session information
session = {
    'experiment' : 'THF',
    'subject' : 'wv001',
    'session' : 2    
}
```

## Load Data

```python
# Load NWB file
nwbfile = load_nwbfile(session, data_folder)
```

## Plot response places

```python
# Collect chest positions
chest_xs, chest_ys = nwbfile.acquisition['stimuli']['chest_positions'].data[:].T
```

```python
# Get recall selector positon data
recall_pos = nwbfile.acquisition['recall']['selector'].data[:].T
```

```python
# Plot the positions of the recall selector, overlaid with chest positions
plot_positions(recall_pos, figsize=(5, 7), 
               landmarks = {'positions' : np.array([chest_xs, chest_ys]),
                            'color' : 'green', 'ms' : 10, 'alpha' : 0.75})
```
